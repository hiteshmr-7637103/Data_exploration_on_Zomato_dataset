# Zomato Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("zomato.csv",encoding="latin-1")
print(df.shape)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
country = pd.read_excel("Country-Code.xlsx")
print("shape:",country.shape)
df = df.merge(country,on='Country Code',how='left')
df.shape
```

```python
country.isnull().sum()
```

```python
df['Country'].value_counts()
```

plotting the top most countries where zomato is used

```python
country_index = df['Country'].value_counts().index
country_value = df['Country'].value_counts().values

plt.pie(country_value,labels=country_index,autopct='%.2f%%')
plt.show()
```

Since the data is clustered i want to list only the top three countries

```python
plt.pie(country_value[:3],labels=country_index[:3],autopct='%.2f%%')
```

```python
ratings = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Aggregate rating',y='Rating count',data = ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','#006400'])
plt.show()
```

1) number of non rated restaurents are high
2) maximum number of ratings lies in the range 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','#006400'])
```

### Find the countries with zero rating

### Find the countries and their currencies

```python
# country with the currency 
df.groupby(['Country','Currency']).size().reset_index().rename(columns={0:"count"})
```

### Countries with online delivery option

```python
df[df['Has Online delivery']=='Yes'].groupby(['Country']).size().reset_index().rename(columns={0:"Count"})
```
